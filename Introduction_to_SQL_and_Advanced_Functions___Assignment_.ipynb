Ques 1 :-  Explain the fundamental differences between DDL, DML, and DQL commands in SQL. Provide one example for each type of command.

Answer:- SQL commands are divided into different categories based on their functions, with the three main types being DDL (Data Definition Language), DML (Data Manipulation Language), and DQL (Data Query Language). DDL commands are used to define and modify the structure of a database, such as creating, altering, or deleting tables. For example, the `CREATE TABLE Students (StudentID INT, Name VARCHAR(50), Age INT);` command creates a new table named **Students**. DML commands are used to manipulate the data stored in database tables by inserting, updating, or deleting records. For example, `INSERT INTO Students (StudentID, Name, Age) VALUES (1, 'John', 20);` adds a new record to the **Students** table. DQL commands are used to retrieve data from the database using the `SELECT` statement. For example, `SELECT * FROM Students;` displays all the records stored in the **Students** table. Thus, DDL focuses on the database structure, DML manages the data within the tables, and DQL is used to query and retrieve data.


Ques 2 :-  What is the purpose of SQL constraints? Name and describe three common types of constraints, providing a simple scenario where each would be useful.  

Answer:- SQL constraints are rules applied to database tables to ensure the accuracy, consistency, and integrity of the data stored in them. They help prevent invalid or duplicate data from being entered into the database. One common constraint is the **PRIMARY KEY**, which uniquely identifies each record in a table and does not allow duplicate or NULL values. For example, the `StudentID` column in a **Students** table can be set as the primary key so that every student has a unique identification number. Another important constraint is the **FOREIGN KEY**, which establishes a relationship between two tables and ensures that the value in one table exists in another. For instance, a `DepartmentID` in an **Employees** table can reference the `DepartmentID` in a **Departments** table, ensuring that every employee belongs to a valid department. The **NOT NULL** constraint is used to ensure that a column cannot contain NULL values. For example, the `Name` column in a **Students** table can be defined as NOT NULL so that every student record must include a name. These constraints help maintain reliable, accurate, and consistent data within a database.


Ques 3 :- Explain the difference between LIMIT and OFFSET clauses in SQL. How would you use them together to retrieve the third page of results, assuming each page has 10 records?

Answer:- The **LIMIT** and **OFFSET** clauses in SQL are used together to control the number of records returned by a query, making them especially useful for pagination. The **LIMIT** clause specifies the maximum number of rows to retrieve, while the **OFFSET** clause specifies how many rows to skip before starting to return results. For example, if each page displays **10 records**, the **third page** should skip the first **20 records** (pages 1 and 2) and then return the next **10 records**. This can be achieved using the query: `SELECT * FROM Employees LIMIT 10 OFFSET 20;`. Here, `LIMIT 10` returns 10 records, and `OFFSET 20` skips the first 20 records, displaying records **21 to 30**, which represent the third page of results.


Ques 4 :- What is a Common Table Expression (CTE) in SQL, and what are its main benefits? Provide a simple SQL example demonstrating its usage.

Answer:- A **Common Table Expression (CTE)** is a temporary named result set in SQL that is created using the `WITH` clause and exists only for the duration of a single query. It is mainly used to simplify complex SQL queries by breaking them into smaller, more readable parts. CTEs improve code readability, make queries easier to maintain, and can also be used for recursive queries, such as working with hierarchical data. For example, to display employees whose salary is greater than 50,000, a CTE can first store all employees with salaries and then retrieve only the required records. The SQL query is: `WITH HighSalaryEmployees AS (SELECT EmployeeID, Name, Salary FROM Employees WHERE Salary > 50000) SELECT * FROM HighSalaryEmployees;`. In this example, `HighSalaryEmployees` is the CTE that temporarily stores employees earning more than 50,000, and the main query retrieves all records from it. This approach makes the query more organized and easier to understand compared to using nested subqueries.


Ques 5 :-  Describe the concept of SQL Normalization and its primary goals. Briefly explain the first three normal forms (1NF, 2NF, 3NF).

Answer:- SQL **Normalization** is the process of organizing data in a database to reduce redundancy and improve data integrity. Its primary goals are to eliminate duplicate data, ensure data consistency, minimize data modification anomalies (such as update, insert, and delete anomalies), and make the database more efficient and easier to maintain. Normalization is achieved by dividing large tables into smaller, related tables and establishing relationships between them using keys. The **First Normal Form (1NF)** requires that each table contains only atomic (indivisible) values, each column contains a single value, and each record is unique. The **Second Normal Form (2NF)** builds on 1NF by ensuring that all non-key attributes are fully dependent on the entire primary key, thereby removing partial dependencies in tables with composite keys. The **Third Normal Form (3NF)** builds on 2NF by removing transitive dependencies, meaning that non-key attributes should depend only on the primary key and not on other non-key attributes. Together, these normal forms help create a well-structured database that is efficient, consistent, and easy to manage.

Ques 9 :-  Write a SQL query that uses a Common Table Expression (CTE) and a Window Function (specifically ROW_NUMBER() or RANK()) to display the CategoryName, ProductName, and Price for the top 2 most expensive products in each CategoryName.

Answer:- WITH ProductRank AS (
    SELECT
        c.CategoryName,
        p.ProductName,
        p.Price,
        ROW_NUMBER() OVER (
            PARTITION BY c.CategoryName
            ORDER BY p.Price DESC
        ) AS RowNum
    FROM Products p
    INNER JOIN Categories c
        ON p.CategoryID = c.CategoryID
)

SELECT
    CategoryName,
    ProductName,
    Price
FROM ProductRank
WHERE RowNum <= 2
ORDER BY CategoryName, Price DESC;

Answer:- SELECT
    c.customer_id,
    CONCAT(c.first_name, ' ', c.last_name) AS CustomerName,
    c.email,

JOIN payment p
ON c.customer_id = p.customer_id
