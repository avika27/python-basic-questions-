Ques 1 :-  What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

Answer:- In a **Convolutional Neural Network (CNN)**, **filters** (also called kernels) and **feature maps** play a crucial role in extracting and learning important patterns from input images. A **filter** is a small matrix of weights that slides over the input image and performs a convolution operation to detect specific features such as edges, corners, textures, or shapes. During training, the CNN automatically learns the optimal values of these filters to recognize meaningful patterns. The output produced after applying a filter is called a **feature map** (or activation map). A feature map highlights the locations where the corresponding feature is detected in the input image. Multiple filters generate multiple feature maps, allowing the CNN to capture different aspects of the image. As the network becomes deeper, these feature maps represent increasingly complex features, progressing from simple edges to object parts and finally complete objects. Thus, filters are responsible for feature extraction, while feature maps store and represent the detected features for further processing by the network.


Ques 2 :- Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?

Answer:- In **Convolutional Neural Networks (CNNs)**, **padding** and **stride** are two important parameters that control how the convolution operation is performed and determine the size of the output feature maps. **Padding** involves adding extra rows and columns of zeros around the border of the input image before applying the convolution. Its main purpose is to preserve the spatial dimensions of the input, prevent the loss of information at the edges, and allow filters to process border pixels effectively. Without padding, the output feature map becomes smaller after each convolution. **Stride** refers to the number of pixels by which the filter moves across the input image during the convolution operation. A stride of **1** means the filter moves one pixel at a time, producing a larger feature map with more detailed information, while a stride greater than **1** moves the filter multiple pixels at a time, reducing the size of the output feature map and decreasing computational cost. Therefore, increasing the padding generally helps maintain or increase the output dimensions, whereas increasing the stride reduces the dimensions of the feature maps. Both parameters are carefully chosen to balance feature extraction, computational efficiency, and the preservation of important image information.


Ques 3 :-  Define receptive field in the context of CNNs. Why is it important for deep architectures?

Answer:- In the context of **Convolutional Neural Networks (CNNs)**, the **receptive field** refers to the region of the input image that influences the value of a particular neuron in a feature map. In other words, it is the area of the input image that a neuron "sees" when making its computation. In the early layers of a CNN, the receptive field is small and captures simple features such as edges, corners, and textures. As the network becomes deeper, the receptive field increases because each layer combines information from the previous layers, allowing neurons to capture larger and more complex patterns such as object parts and entire objects. The receptive field is important in deep architectures because it enables the network to learn both local and global features of an image. A larger receptive field helps the model understand the overall context of the image, improves object recognition, and enhances performance on complex computer vision tasks such as image classification, object detection, and semantic segmentation. Thus, the receptive field is a key factor in enabling CNNs to extract meaningful hierarchical features from images.


Ques 4 :-  Discuss how filter size and stride influence the number of parameters in a CNN.

Answer:- In a **Convolutional Neural Network (CNN)**, the **filter size** and **stride** influence how features are extracted from an input image, but they affect the number of parameters differently. The **filter size** directly determines the number of learnable parameters because each filter contains weights that are updated during training. A larger filter, such as **5 × 5**, has more parameters than a smaller filter like **3 × 3**, resulting in higher computational cost and increased memory usage. On the other hand, **stride** specifies how many pixels the filter moves across the input image during convolution. Changing the stride does **not** change the number of learnable parameters because the filter weights remain the same. However, a larger stride reduces the size of the output feature maps, which decreases the number of computations and memory required during processing. Therefore, while the **filter size directly affects the number of parameters**, the **stride mainly affects the output dimensions and computational efficiency**, helping balance model complexity and performance.


Ques 5:- : Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.

Answer:- **LeNet**, **AlexNet**, and **VGG** are three important Convolutional Neural Network (CNN) architectures that differ in their depth, filter sizes, and overall performance. **LeNet**, introduced by Yann LeCun, is one of the earliest CNN models and consists of about **5–7 layers** with relatively small filters, making it suitable for simple tasks such as handwritten digit recognition. **AlexNet**, developed in 2012, is much deeper with **8 layers** (5 convolutional and 3 fully connected layers) and uses larger filters in the first layer (11 × 11) followed by smaller filters. It introduced the use of **ReLU activation**, **dropout**, and **GPU training**, significantly improving image classification performance and winning the ImageNet competition. **VGG**, introduced by the Visual Geometry Group, is a much deeper architecture with **16 or 19 layers** and uses only **3 × 3 filters** throughout the network. Although VGG has a large number of parameters and requires more computational resources, its uniform architecture allows it to learn more complex features and achieve higher accuracy than earlier models. In summary, **LeNet** is simple and lightweight, **AlexNet** is deeper and more powerful with improved training techniques, and **VGG** is the deepest among the three, using small filters to achieve superior performance at the cost of increased computational complexity.


```python
# Ques 6 :- Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation. (Include your Python code and output in the code box below.)

# Import required libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.datasets import mnist

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape the images to (28, 28, 1)
x_train = x_train.reshape((60000, 28, 28, 1)).astype("float32")
x_test = x_test.reshape((10000, 28, 28, 1)).astype("float32")

# Normalize pixel values to range [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Create the CNN model
model = Sequential()

# First Convolutional Layer
model.add(Conv2D(filters=32,
                 kernel_size=(3,3),
                 activation='relu',
                 input_shape=(28,28,1)))

# First Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Second Convolutional Layer
model.add(Conv2D(filters=64,
                 kernel_size=(3,3),
                 activation='relu'))

# Second Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))

# Flatten the output
model.add(Flatten())

# Fully Connected Hidden Layer
model.add(Dense(128, activation='relu'))

# Output Layer
model.add(Dense(10, activation='softmax'))

# Display model architecture
model.summary()

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train the model
history = model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=1)

# Print the results
print("\nTest Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# Ques 8 :-  Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.

# Import required libraries
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Check device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Data Preprocessing
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Load MNIST Dataset
train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

# Data Loaders
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# CNN Model Definition
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 5 * 5, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x

# Create Model
model = CNN().to(device)

# Loss Function and Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training Loop
epochs = 5

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader):.4f}")

# Model Evaluation
model.eval()

correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:

        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)

        _, predicted = torch.max(outputs.data, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total

print(f"\nTest Accuracy: {accuracy:.2f}%")
```

```python
# Ques 9 :- Given a custom image dataset stored in a local directory, write code using Keras ImageDataGenerator to preprocess and train a CNN model.

# Import required libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Path to the dataset
train_dir = "dataset/train"
test_dir = "dataset/test"

# Image Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Load Training Data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

# Load Validation Data
validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Load Test Data
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Build CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),

    Dense(128, activation='relu'),
    Dense(train_generator.num_classes, activation='softmax')
])

# Display Model Summary
model.summary()

# Compile Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train Model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10
)

# Evaluate Model
test_loss, test_accuracy = model.evaluate(test_generator)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# Ques 10 :-  You are working on a web application for a medical imaging startup. Your task is to build and deploy a CNN model that classifies chest X-ray images into “Normal” and “Pneumonia” categories. Describe your end-to-end approach–from data preparation and model training to deploying the model as a web app using Streamlit.


import os
import numpy as np
import streamlit as st
from PIL import Image
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ============================
# Train Model (Runs Once)
# ============================

MODEL_PATH = "pneumonia_model.keras"

if not os.path.exists(MODEL_PATH):

    train_dir = "dataset/train"
    val_dir = "dataset/val"
    test_dir = "dataset/test"

    # Image Preprocessing
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=15,
        zoom_range=0.2,
        horizontal_flip=True
    )

    test_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224,224),
        batch_size=32,
        class_mode='binary'
    )

    validation_generator = test_datagen.flow_from_directory(
        val_dir,
        target_size=(224,224),
        batch_size=32,
        class_mode='binary'
    )

    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(224,224),
        batch_size=32,
        class_mode='binary',
        shuffle=False
    )

    # CNN Model
    model = Sequential([
        Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
        MaxPooling2D(2,2),

        Conv2D(64, (3,3), activation='relu'),
        MaxPooling2D(2,2),

        Conv2D(128, (3,3), activation='relu'),
        MaxPooling2D(2,2),

        Flatten(),

        Dense(128, activation='relu'),
        Dropout(0.5),

        Dense(1, activation='sigmoid')
    ])

    # Compile Model
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Train Model
    model.fit(
        train_generator,
        validation_data=validation_generator,
        epochs=10
    )

    # Evaluate
    loss, accuracy = model.evaluate(test_generator)
    print("Test Accuracy:", accuracy)

    # Save Model
    model.save(MODEL_PATH)

# ============================
# Load Saved Model
# ============================

model = tf.keras.models.load_model(MODEL_PATH)

# ============================
# Streamlit Web Application
# ============================

st.title("Chest X-Ray Classification")
st.write("Upload a Chest X-ray image to predict whether it is Normal or Pneumonia.")

uploaded_file = st.file_uploader(
    "Choose a Chest X-ray Image",
    type=["jpg", "jpeg", "png"]
)

if uploaded_file is not None:

    image = Image.open(uploaded_file).convert("RGB")

    st.image(image, caption="Uploaded Image", use_container_width=True)

    # Preprocess Image
    img = image.resize((224,224))
    img = np.array(img).astype("float32") / 255.0
    img = np.expand_dims(img, axis=0)

    # Prediction
    prediction = model.predict(img)

    probability = prediction[0][0]

    if probability > 0.5:
        st.error("Prediction: Pneumonia")
        st.write(f"Confidence: {probability*100:.2f}%")
    else:
        st.success("Prediction: Normal")
        st.write(f"Confidence: {(1-probability)*100:.2f}%")
```
