Ques 13:-  What is a property decorator in Python?

Answer:-A property decorator in Python allows you to access class methods as attributes without using parentheses ().
It is defined using the @property decorator.

With @property, a method behaves like a read-only attribute, and we can also define setter and deleter methods using @<property_name>.setter and @<property_name>.deleter to modify or delete the property’s value.

This helps in implementing encapsulation by controlling how attributes are accessed and modified.

eg:- class Circle:

    def __init__(self, radius):
        self._radius = radius

    @property
    def area(self):
        return 3.14 * self._radius * self._radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        self._radius = value


obj1 = Circle(5)

print(obj1.area)      
obj1.radius = 10      
print(obj1.area)

Ques 17:-  What is the difference between a class variable and an instance variable?

Answer:-
A class variable is a variable that is shared among all objects of a class. It is defined inside the class but outside any methods, meaning there is only one copy of it that all instances share. Ques 20:-What is the significance of the ‘super()’ function in Python?

Answer:-the super() function is used to call methods of a parent (or base) class inside a child class. It allows you to access the parent class’s features without directly naming it, which makes the code cleaner and helps in multiple inheritance. The main significance of super() is that it helps reuse the parent class's methods and variables, avoid code duplication, and maintain proper method order (especially in multiple inheritance using MRO – Method Resolution Order). For example:

class Parent:

    def __init__(self):
        print("Parent class constructor")

class Child(Parent):

    def __init__(self):
        super().__init__()  # Calling parent class constructor
        print("Child class constructor")

obj = Child()


Here, when we create an object of Child, super() calls the Parent class constructor first, making sure the parent functionality runs before the child's. Thus, super() ensures better code reuse and proper inheritance behavior.

Ques 22:-What is the difference between @staticmethod and @classmethod in Python?

Answer:-@staticmethod and @classmethod are decorators used to define methods inside a class, but they work differently.

A @staticmethod is a method that does not take the self or cls parameter. It behaves like a normal function but is placed inside a class for logical grouping. It cannot access or modify class or object (instance) data. It is used when the method performs a task that is related to the class but does not depend on class attributes or instance attributes.

A @classmethod, on the other hand, takes cls as its first parameter (instead of self). It can access and modify class-level data and is useful when you want to work with class variables or create alternate constructors.

```python
#Ques 5:-Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
#balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn {amount}")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print(f"Current Balance: {self.__balance}")


acc = BankAccount(1000)

acc.deposit(500)
acc.withdraw(200)
acc.check_balance()
```

```python
#Ques 14:-Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
#and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, hours, rate):
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate


class Manager(Employee):
    def __init__(self, hours, rate, bonus):
        super().__init__(hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus



emp = Employee(40, 200)
print("Employee Salary:", emp.calculate_salary())

mgr = Manager(40, 200, 5000)
print("Manager Salary:", mgr.calculate_salary())
```

```python
#Ques 16:- Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
#implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass


class Cow(Animal):
    def sound(self):
        print("Cow says: Moo")


class Sheep(Animal):
    def sound(self):
        print("Sheep says: Baa")


cow = Cow()
sheep = Sheep()

cow.sound()
sheep.sound()

```

```python
#Ques 18:-Create a class House with attributes address and price. Create a derived class Mansion that adds an
#attribute number_of_rooms.

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)   # call parent constructor
        self.number_of_rooms = number_of_rooms

    def display(self):
        print("Address:", self.address)
        print("Price:", self.price)
        print("Number of Rooms:", self.number_of_rooms)



obj = Mansion("Delhi", 50000000, 10)
obj.display()
```
