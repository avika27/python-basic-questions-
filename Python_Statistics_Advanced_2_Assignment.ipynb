Ques 1 :-What is hypothesis testing in statistics?

Answer:- Hypothesis testing is a statistical procedure used to evaluate assumptions or claims about a population using sample data. It involves formulating a null hypothesis, which states that there is no effect or difference, and an alternative hypothesis, which indicates the presence of an effect or difference. Data is then collected and analyzed using appropriate statistical tests. A significance level is chosen to determine the acceptable risk of making an incorrect decision. Based on the calculated test statistic or p-value, a decision is made to either reject or fail to reject the null hypothesis, helping researchers draw objective and reliable conclusions.

Ques 2:-  What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:-The null hypothesis (H₀) is a statement that assumes there is no effect, no difference, or no relationship between variables in a population. It represents the default position that any observed change in the sample is due to random chance. In contrast, the alternative hypothesis (H₁ or Hₐ) states that there is a significant effect, difference, or relationship and contradicts the null hypothesis. While the null hypothesis is tested directly using statistical methods, the alternative hypothesis is accepted only when sufficient evidence is found to reject the null hypothesis.


Ques 3 :-Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

Answer:-The significance level (α) in hypothesis testing represents the maximum probability of making a Type I error, which means rejecting a true null hypothesis. It is chosen by the researcher before conducting the test, commonly set at values like 0.05 or 0.01. The significance level acts as a decision threshold for evaluating the p-value obtained from sample data. If the p-value is less than or equal to α, the null hypothesis is rejected; if it is greater than α, the null hypothesis is not rejected. Thus, the significance level controls the risk of error and plays a crucial role in determining the outcome of a hypothesis test.

Ques 4 :-What are Type I and Type II errors? Give examples of each.

Answer:-In hypothesis testing, Type I and Type II errors are decision errors that occur due to uncertainty in sample data.

A Type I error occurs when a true null hypothesis is rejected. In simple terms, it is a false positive.
Example: A medicine is actually not effective, but the test results lead researchers to conclude that it works.

A Type II error occurs when a false null hypothesis is not rejected. This is known as a false negative.
Example: A medicine is actually effective, but the test results fail to detect its effect and conclude that it does not work.

Ques 5 :- What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer:-The Z-test and T-test are both statistical tests used to determine whether there is a significant difference between sample and population parameters, but they differ based on sample size and knowledge of population variability. A Z-test is used when the sample size is large (n ≥ 30) and the population standard deviation is known, or when the data follows a normal distribution. In contrast, a T-test is used when the sample size is small (n < 30) and the population standard deviation is unknown, in which case the sample standard deviation is used. The T-test is more flexible and commonly applied in practical research situations, especially with limited data.

```python
# Ques 6:- Write a Python program to generate a binomial distribution with n=10 and
#p=0.5, then plot its histogram.


import random
import matplotlib.pyplot as plt

n = 10
p = 0.5
num_samples = 1000


data = []
for _ in range(num_samples):
    successes = 0
    for _ in range(n):
        if random.random() < p:
            successes += 1
    data.append(successes)


plt.hist(data, bins=range(0, n+2), edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()


print("Sample output (first 10 values):")
print(data[:10])
```

```python
#Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
 #                       50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
  #                      50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
   #                     50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import math
from statistics import mean, stdev
from scipy.stats import norm


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu = 50


x_bar = mean(sample_data)
s = stdev(sample_data)
n = len(sample_data)


z = (x_bar - mu) / (s / math.sqrt(n))


p_value = 2 * (1 - norm.cdf(abs(z)))

print("Sample Mean:", x_bar)
print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
# Ques 8 :- Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm

np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)

margin_of_error = z_critical * (sample_std / math.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error


print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))


plt.hist(data, bins=10)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Simulated Normal Distribution Data")
plt.show()
```

```python
# Ques 9:-: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt


def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores


data = [45, 48, 50, 52, 55, 49, 51, 53, 47, 54]


z_scores = calculate_z_scores(data)


print("Z-scores:")
print(z_scores)

plt.hist(z_scores, bins=8)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Standardized Data (Z-scores)")
plt.show()
```
