Ques 1 :- What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

Answer:-Descriptive statistics and inferential statistics are two main branches of statistics that serve different purposes. Descriptive statistics is concerned with summarizing and presenting data in a simple and meaningful way. It describes the main features of a dataset using measures such as mean, median, mode, range, and standard deviation, and also uses tables, graphs, and charts for clear representation. It does not go beyond the given data and does not make predictions or generalizations. For example, if the average marks of a class of 40 students is calculated as 68, descriptive statistics only tells us about the performance of that particular class. On the other hand, inferential statistics is used to draw conclusions, make predictions, or generalize results about a larger population based on a sample of data. It involves techniques like hypothesis testing, confidence intervals, t-tests, and regression analysis. For example, if a sample of 200 students from a university is studied and their average study time is found to be 4 hours per day, inferential statistics helps in concluding that most students in the entire university may study around 4 hours per day. In short, descriptive statistics explains what the data shows, while inferential statistics helps in making decisions and predictions beyond the available data.

Ques 2 :-  What is sampling in statistics? Explain the differences between random
and stratified sampling.

Answer:- ampling in statistics is the process of selecting a small group of individuals or observations (called a sample) from a larger group (called a population) in order to study it and draw conclusions about the whole population. Since studying an entire population is often time-consuming, costly, or impractical, sampling allows researchers to save time and resources while still obtaining reliable and meaningful results.

Random sampling is a method in which every member of the population has an equal and independent chance of being selected in the sample. The selection is done purely by chance, such as using a lottery method or random number tables. This method reduces selection bias and is easy to apply, but it may not properly represent all subgroups of the population. For example, if 50 students are randomly selected from a college of 1,000 students, each student has an equal chance of being chosen, but some departments may be over-represented or under-represented.

Stratified sampling, on the other hand, involves dividing the population into smaller homogeneous groups called strata based on certain characteristics such as age, gender, or department. A random sample is then drawn from each stratum, usually in proportion to its size in the population. This method ensures that all important subgroups are adequately represented in the sample and generally provides more accurate results. For example, in a college with science, arts, and commerce students, stratified sampling would involve selecting students from each stream in proportion to their actual numbers. In summary, random sampling relies purely on chance, while stratified sampling ensures representation from all key subgroups of the population.

Ques 3 :- Define mean, median, and mode. Explain why these measures of central
tendency are important.

Answer:- Mean, median, and mode are the three main measures of central tendency used in statistics to describe the central or typical value of a dataset. The mean is the arithmetic average of all observations and is calculated by adding all values and dividing by the total number of observations. For example, if the values are 2, 4, and 6, the mean is (2 + 4 + 6) / 3 = 4. The median is the middle value of an ordered dataset; when the number of observations is even, it is the average of the two middle values. For example, in the dataset 3, 5, 7, 9, and 11, the median is 7. The mode is the value that occurs most frequently in a dataset; a dataset may have one mode, more than one mode, or no mode at all. These measures of central tendency are important because they provide a simple summary of large datasets, help in comparing different datasets, and give a quick understanding of the typical value around which the data is concentrated.

Ques 4 :-  Explain skewness and kurtosis. What does a positive skew imply about the
data?

Answer:- Skewness and kurtosis are statistical measures that describe the shape of a data distribution. Skewness refers to the degree of asymmetry in a distribution around its mean. If the data is symmetrically distributed, skewness is zero. When the distribution has a longer tail on the right side, it is called positive (right) skewness, whereas a longer tail on the left side indicates negative (left) skewness. For example, income distribution is usually positively skewed because most people earn moderate incomes while a few earn very high incomes. A positive skew implies that the data has more lower values with some extreme high values, and typically the relationship between mean, median, and mode is Mean > Median > Mode.

Kurtosis, on the other hand, measures the peakedness or flatness of a distribution compared to a normal distribution. High kurtosis (leptokurtic) indicates a sharp peak with heavy tails, meaning more extreme values or outliers, while low kurtosis (platykurtic) shows a flatter distribution with fewer extreme values. Normal kurtosis (mesokurtic) represents a balanced distribution similar to the normal curve. Together, skewness and kurtosis help in understanding the shape, spread, and presence of outliers in data, which is important for choosing appropriate statistical methods and making accurate interpretations.

```python
# Ques 5 :  Implement a Python program to compute the mean, median, and mode of
#a given list of numbers.
#numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Mean
mean_value = statistics.mean(numbers)

# Median
median_value = statistics.median(numbers)

# Mode
mode_value = statistics.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
# Ques 6 :- Compute the covariance and correlation coefficient between the following
#two datasets provided as lists in Python:
#list_x = [10, 20, 30, 40, 50]
#list_y = [15, 25, 35, 45, 60]

import numpy as np

list_x = np.array([10, 20, 30, 40, 50])
list_y = np.array([15, 25, 35, 45, 60])

# Covariance
covariance = np.cov(list_x, list_y)[0][1]

# Correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
# Ques 7 :-Write a Python script to draw a boxplot for the following numeric list and
#identify its outliers. Explain the result:
#data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Given Data")
plt.ylabel("Values")
plt.show()

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Outliers:", outliers)
```

```python
# Ques 8:- You are working as a data analyst in an e-commerce company. The
#marketing team wants to know if there is a relationship between advertising spend and
#daily sales.
#● Explain how you would use covariance and correlation to explore this
#relationship.
#● Write Python code to compute the correlation between the two lists:
#advertising_spend = [200, 250, 300, 400, 500]
#daily_sales = [2200, 2450, 2750, 3200, 4000]

#Answer:- import numpy as np

# To understand whether advertising spend is related to daily sales,
# covariance and correlation are used as exploratory statistical tools.
# Covariance helps identify the direction of the relationship between
# the two variables. If the covariance is positive, it indicates that as
# advertising spend increases, daily sales also tend to increase; if it
# is negative, sales tend to decrease as advertising spend increases.
# However, covariance does not tell us how strong the relationship is
# because its value depends on the scale of the data.
#
# To overcome this limitation, the correlation coefficient is used.
# Correlation standardizes the relationship and gives a value between
# -1 and +1. A correlation close to +1 indicates a strong positive
# relationship (higher advertising spend leads to higher sales),
# close to -1 indicates a strong negative relationship, and close to
# 0 indicates little or no linear relationship. Together, covariance
# tells the direction, and correlation tells the strength of the relationship.

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to NumPy arrays
ad_spend = np.array(advertising_spend)
sales = np.array(daily_sales)

# Compute correlation coefficient
correlation = np.corrcoef(ad_spend, sales)[0][1]

print("Correlation Coefficient:", correlation)
```
