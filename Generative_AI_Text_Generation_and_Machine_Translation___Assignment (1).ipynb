Overall, Generative AI increases productivity, reduces manual effort, enhances creativity, and enables organizations to automate complex tasks across a wide range of industries.

Ques 2 :- Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?

Answer:- Probabilistic modeling is the foundation of many generative models because it represents the probability distribution of the data. Instead of simply recognizing patterns, these models learn how the data is generated by estimating the joint probability distribution, P(X,Y), or the data distribution P(X). Once this distribution is learned, the model can generate new samples that are similar to the training data, estimate missing values, and handle uncertainty. Examples of probabilistic generative models include Variational Autoencoder (VAEs), Gaussian Mixture Model (GMMs), and Hidden Markov Model (HMMs).

The main difference between generative and discriminative models lies in what they learn. Generative models learn how the data is produced by modeling its probability distribution, allowing them to generate new data samples and perform classification. In contrast, discriminative models learn the decision boundary between different classes by directly estimating the conditional probability P(Yв€ЈX). Their primary goal is accurate prediction rather than data generation.

For example, a generative model can learn the characteristics of handwritten digits and create new realistic digit images, while a discriminative model only learns to classify an input image as a specific digit (such as 0вЂ“9). Generative models are widely used for image synthesis, text generation, data augmentation, and anomaly detection, whereas discriminative models are mainly used for classification and regression tasks. In general, discriminative models often achieve higher classification accuracy, while generative models provide greater flexibility by learning the underlying structure of the data and generating new examples.

Ques 3 :- What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?

Answer:- A standard Autoencoder (AE) is a neural network that learns to compress input text into a compact latent representation and then reconstruct the original text from that representation. It consists of an encoder, which converts the input into a fixed latent vector, and a decoder, which reconstructs the original text. The main objective of an autoencoder is to minimize the reconstruction error, making it useful for tasks such as feature extraction, dimensionality reduction, and text compression. However, because the latent representation is deterministic and not organized for sampling, an autoencoder cannot effectively generate new and diverse text.

In contrast, a Variational Autoencoder (VAE) is a probabilistic generative model that learns a distribution over the latent space rather than a single fixed representation. Instead of producing one latent vector, the encoder predicts the mean and variance of a probability distribution, from which a latent vector is sampled before being passed to the decoder. This creates a smooth and continuous latent space that enables the generation of new, realistic, and diverse text samples. As a result, VAEs are widely used for text generation, sentence completion, paraphrasing, and data augmentation. In summary, while autoencoders are mainly designed to reconstruct existing text, variational autoencoders are specifically designed to generate new text by learning the underlying probability distribution of the data.


Ques 4 :- Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

Answer:- In Neural Machine Translation (NMT), the attention mechanism allows the model to focus on the most relevant words in the source sentence while generating each word in the translated sentence. In a traditional encoder-decoder architecture, the encoder converts the entire input sentence into a single fixed-length context vector, which can lose important information, especially for long sentences. The attention mechanism overcomes this limitation by enabling the decoder to access all the encoder's hidden states instead of relying on a single context vector. During the generation of each target word, the decoder calculates attention weights that indicate the importance of each source word and creates a context vector as a weighted combination of the encoder's hidden states. This context vector is then used along with the decoder's current state to predict the next word in the translated sentence.

Attention mechanisms are critical because they significantly improve translation quality, particularly for long and complex sentences. They help the model capture long-range dependencies, maintain the correct context, and accurately align words between the source and target languages. By focusing on the most relevant parts of the input at each decoding step, attention reduces information loss and produces more fluent and accurate translations. Furthermore, attention forms the foundation of modern transformer-based models, which have achieved state-of-the-art performance in machine translation and many other natural language processing tasks.

```python
# Ques 6 :-  Use the following small text dataset to train a simple Variational
#Autoencoder (VAE) for text reconstruction:
#["The sky is blue", "The sun is bright", "The grass is green",
#"The night is dark", "The stars are shining"]
#1. Preprocess the data (tokenize and pad the sequences).
#2. Build a basic VAE model for text reconstruction.
#3. Train the model and show how it reconstructs or generates similar sentences.
#Include your code, explanation, and sample outputs


import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Layer, Input, Embedding, LSTM, Dense, RepeatVector, TimeDistributed
from tensorflow.keras.models import Model

# -----------------------------
# Dataset
# -----------------------------
sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

tokenizer = Tokenizer(lower=True)
tokenizer.fit_on_texts(sentences)

seqs = tokenizer.texts_to_sequences(sentences)
max_len = max(len(s) for s in seqs)
X = pad_sequences(seqs, maxlen=max_len, padding="post")

vocab_size = len(tokenizer.word_index) + 1

print("Vocabulary Size:", vocab_size)
print("Max Length:", max_len)
print(X)

embedding_dim = 16
latent_dim = 8

# -----------------------------
# Sampling Layer
# -----------------------------
class Sampling(Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        eps = tf.random.normal(shape=tf.shape(z_mean))
        return z_mean + tf.exp(0.5 * z_log_var) * eps

# -----------------------------
# Encoder
# -----------------------------
encoder_inputs = Input(shape=(max_len,))
x = Embedding(vocab_size, embedding_dim, mask_zero=True)(encoder_inputs)
x = LSTM(32)(x)

z_mean = Dense(latent_dim)(x)
z_log_var = Dense(latent_dim)(x)
z = Sampling()([z_mean, z_log_var])

encoder = Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# -----------------------------
# Decoder
# -----------------------------
latent_inputs = Input(shape=(latent_dim,))
x = RepeatVector(max_len)(latent_inputs)
x = LSTM(32, return_sequences=True)(x)
decoder_outputs = TimeDistributed(Dense(vocab_size, activation="softmax"))(x)

decoder = Model(latent_inputs, decoder_outputs, name="decoder")

# -----------------------------
# VAE Model
# -----------------------------
class VAE(Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = tf.keras.metrics.Mean(name="loss")
        self.rec_loss_tracker = tf.keras.metrics.Mean(name="reconstruction_loss")
        self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

    @property
    def metrics(self):
        return [
            self.total_loss_tracker,
            self.rec_loss_tracker,
            self.kl_loss_tracker,
        ]

    def train_step(self, data):
        if isinstance(data, tuple):
            data = data[0]

        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data, training=True)
            reconstruction = self.decoder(z, training=True)

            rec_loss = tf.reduce_mean(
                tf.reduce_sum(
                    tf.keras.losses.sparse_categorical_crossentropy(
                        data, reconstruction
                    ),
                    axis=1,
                )
            )

            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(
                    1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var),
                    axis=1,
                )
            )

            total_loss = rec_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.rec_loss_tracker.update_state(rec_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.total_loss_tracker.result(),
            "reconstruction_loss": self.rec_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

vae = VAE(encoder, decoder)
vae.compile(optimizer="adam")

print("\nTraining...\n")
vae.fit(X, epochs=300, batch_size=2, verbose=1)

# -----------------------------
# Reconstruction
# -----------------------------
print("\nReconstructed Sentences:\n")

_, _, latent = encoder.predict(X, verbose=0)
pred = decoder.predict(latent, verbose=0)

pred_ids = np.argmax(pred, axis=-1)
index_word = tokenizer.index_word

for original, seq in zip(sentences, pred_ids):
    words = [index_word.get(i, "") for i in seq if i != 0]
    print("Original     :", original)
    print("Reconstructed:", " ".join(words))
    print()

# -----------------------------
# Generation
# -----------------------------
print("Generated Sentences:\n")

for _ in range(5):
    z = np.random.normal(size=(1, latent_dim))
    out = decoder.predict(z, verbose=0)
    ids = np.argmax(out, axis=-1)[0]
    words = [index_word.get(i, "") for i in ids if i != 0]
    print("Generated:", " ".join(words))
```

```python
# Ques 8 :- Implement a simple attention-based encoder-decoder model for
#English-to-Spanish translation using Tensorflow or PyTorch.



# Updated Simple Attention-based Encoder-Decoder (TensorFlow/Keras 3)

import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

pairs = [
    ("hello", "hola"),
    ("how are you", "como estas"),
    ("i love you", "te amo"),
    ("good morning", "buenos dias"),
    ("good night", "buenas noches"),
    ("thank you", "gracias"),
    ("see you", "hasta luego"),
    ("what is your name", "como te llamas"),
    ("my name is john", "me llamo john"),
    ("i am happy", "estoy feliz")
]

inp = [x for x, _ in pairs]
tgt = ["START " + y + " END" for _, y in pairs]

inp_tok = Tokenizer(filters='')
tgt_tok = Tokenizer(filters='')

inp_tok.fit_on_texts(inp)
tgt_tok.fit_on_texts(tgt)

enc = inp_tok.texts_to_sequences(inp)
dec = tgt_tok.texts_to_sequences(tgt)

max_enc = max(len(x) for x in enc)
max_dec = max(len(x) for x in dec)

enc = pad_sequences(enc, maxlen=max_enc, padding="post")
dec = pad_sequences(dec, maxlen=max_dec, padding="post")

dec_in = dec[:, :-1]
dec_out = dec[:, 1:]

dataset = tf.data.Dataset.from_tensor_slices((enc, dec_in, dec_out)).batch(2)

enc_vocab = len(inp_tok.word_index) + 1
dec_vocab = len(tgt_tok.word_index) + 1

EMBED = 64
UNITS = 128

class Encoder(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.emb = tf.keras.layers.Embedding(enc_vocab, EMBED)
        self.lstm = tf.keras.layers.LSTM(UNITS, return_sequences=True, return_state=True)

    def call(self, x):
        x = self.emb(x)
        out, h, c = self.lstm(x)
        return out, h, c

class Attention(tf.keras.layers.Layer):
    def __init__(self):
        super().__init__()
        self.W1 = tf.keras.layers.Dense(UNITS)
        self.W2 = tf.keras.layers.Dense(UNITS)
        self.V = tf.keras.layers.Dense(1)

    def call(self, hidden, enc_out):
        hidden = tf.expand_dims(hidden, 1)
        score = self.V(tf.nn.tanh(self.W1(enc_out) + self.W2(hidden)))
        weights = tf.nn.softmax(score, axis=1)
        context = tf.reduce_sum(weights * enc_out, axis=1)
        return context

class Decoder(tf.keras.Model):
    def __init__(self):
        super().__init__()
        self.emb = tf.keras.layers.Embedding(dec_vocab, EMBED)
        self.att = Attention()
        self.lstm = tf.keras.layers.LSTM(UNITS, return_sequences=True, return_state=True)
        self.fc = tf.keras.layers.Dense(dec_vocab)

    def call(self, x, h, c, enc_out):
        context = self.att(h, enc_out)
        x = self.emb(x)
        context = tf.expand_dims(context, 1)
        x = tf.concat([context, x], axis=-1)
        out, h, c = self.lstm(x, initial_state=[h, c])
        out = self.fc(out)
        return out, h, c

encoder = Encoder()
decoder = Decoder()

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction="none")
opt = tf.keras.optimizers.Adam()

def compute_loss(real, pred):
    loss = loss_fn(real, pred)
    mask = tf.cast(tf.not_equal(real, 0), tf.float32)
    return tf.reduce_mean(loss * mask)

@tf.function
def train_step(enc_x, dec_x, target):
    with tf.GradientTape() as tape:
        enc_out, h, c = encoder(enc_x)
        dh, dc = h, c
        outputs = []

        for t in range(dec_x.shape[1]):
            pred, dh, dc = decoder(tf.expand_dims(dec_x[:, t], 1), dh, dc, enc_out)
            outputs.append(pred)

        outputs = tf.concat(outputs, axis=1)
        loss = compute_loss(target, outputs)

    vars = encoder.trainable_variables + decoder.trainable_variables
    grads = tape.gradient(loss, vars)
    opt.apply_gradients(zip(grads, vars))
    return loss

EPOCHS = 300

for e in range(EPOCHS):
    total = 0
    for a, b, c in dataset:
        total += train_step(a, b, c)
    if (e + 1) % 50 == 0:
        print(f"Epoch {e+1}, Loss {float(total):.4f}")

idx2word = {v: k for k, v in tgt_tok.word_index.items()}
start_token = tgt_tok.word_index["start"]
end_token = tgt_tok.word_index["end"]

def translate(sentence):
    seq = inp_tok.texts_to_sequences([sentence])
    seq = pad_sequences(seq, maxlen=max_enc, padding="post")

    enc_out, h, c = encoder(seq)

    token = start_token
    result = []

    for _ in range(max_dec):
        pred, h, c = decoder(tf.constant([[token]]), h, c, enc_out)
        token = int(tf.argmax(pred[0, 0]).numpy())
        word = idx2word.get(token, "")

        if word == "end":
            break
        if word != "start":
            result.append(word)

    return " ".join(result)

print("\nTranslations:\n")
tests = [
    "hello",
    "good morning",
    "thank you",
    "i love you",
    "how are you",
    "my name is john"
]

for t in tests:
    print(f"{t:20s} -> {translate(t)}")
```

```python
# Ques 9 :- Use the following short poetry dataset to simulate poem generation with a
#pre-trained GPT model:
#["Roses are red, violets are blue,",
#"Sugar is sweet, and so are you.",
#"The moon glows bright in silent skies,",
#"A bird sings where the soft wind sighs."]
#Using this dataset as a reference for poetic structure and language, generate a new 2-4
#line poem using a pre-trained GPT model (such as GPT-2). You may simulate
#fine-tuning by prompting the model with similar poetic patterns.
#Include your code, the prompt used, and the generated poem in your answer.


from transformers import pipeline, set_seed

# Load the pre-trained GPT-2 model
generator = pipeline("text-generation", model="gpt2")

# Set seed for reproducible results
set_seed(42)

# Poetry dataset
poems = [
    "Roses are red, violets are blue,",
    "Sugar is sweet, and so are you.",
    "The moon glows bright in silent skies,",
    "A bird sings where the soft wind sighs."
]

# Prompt simulating fine-tuning
prompt = """
The following are examples of short poems:

Roses are red, violets are blue,
Sugar is sweet, and so are you.

The moon glows bright in silent skies,
A bird sings where the soft wind sighs.

Write a new four-line poem in a similar style:
"""

# Generate poem
result = generator(
    prompt,
    max_new_tokens=50,
    do_sample=True,
    temperature=0.8,
    top_k=50,
    top_p=0.95,
    num_return_sequences=1
)

# Display results
print("Reference Dataset:\n")
for line in poems:
    print(line)

print("\nPrompt Used:\n")
print(prompt)

print("\nGenerated Poem:\n")
print(result[0]["generated_text"])
```
