Ques 1:-What is a random variable in probability theory?

Answer:- A random variable in probability theory is a function that assigns a numerical value to each possible outcome of a random experiment. It allows uncertain outcomes to be represented in numerical form so that probabilities can be measured and analyzed mathematically. The value of a random variable depends on chance, and it can be either discrete, taking specific countable values, or continuous, taking any value within a given range.

Ques 2:-  What are the types of random variables?

Answer:- There are two main types of random variables in probability theory: discrete random variables and continuous random variables. A discrete random variable takes a finite or countable number of values, such as the number of heads obtained in coin tosses or the number on a rolled die. A continuous random variable can take any value within a given interval, such as height, weight, time, or temperature, and its values are not countable.

Ques 3 :- Explain the difference between discrete and continuous distributions.

Answer:-The difference between discrete and continuous distributions is based on the type of values the random variable can take. A discrete distribution describes a random variable that takes countable and distinct values, such as the number of heads in a series of coin tosses or the number obtained when rolling a die. The probabilities of these values are defined using a probability mass function (PMF). In contrast, a continuous distribution describes a random variable that can take any value within a given range, such as a person’s height, weight, or the time taken to complete a task. Continuous distributions use a probability density function (PDF), and the probability of the variable taking any exact single value is zero; instead, probabilities are calculated over intervals.

Ques 4 :-  What is a binomial distribution, and how is it used in probability?

Answer:-A binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent trials of a random experiment, where each trial has only two possible outcomes—success or failure—and the probability of success remains constant in every trial. It is used in probability to model situations such as the number of heads obtained in repeated coin tosses, the number of defective items in a sample, or the number of students who pass an exam. The binomial distribution helps in calculating the probability of obtaining a specific number of successes out of a given number of trials.

Ques 5 :-What is the standard normal distribution, and why is it important?

Answer:-The standard normal distribution is a special type of normal distribution with a mean of 0 and a standard deviation of 1. It is important because any normal distribution can be converted into the standard normal distribution using z-scores, which makes it easy to calculate probabilities and compare values from different datasets. Standard normal tables are widely used in statistics to find probabilities, confidence intervals, and hypothesis testing, making the standard normal distribution a fundamental tool in probability and statistical analysis.

Ques 6 :-What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:-The Central Limit Theorem (CLT) states that when a sufficiently large number of independent samples are taken from any population with a finite mean and variance, the distribution of the sample mean approaches a normal distribution, regardless of the original population’s shape. It is critical in statistics because it allows statisticians to apply normal probability methods to real-world data, enables accurate estimation and hypothesis testing, and forms the theoretical foundation for confidence intervals and many statistical inference techniques.

Ques 7:-What is the significance of confidence intervals in statistical analysis?

Answer:-The significance of confidence intervals in statistical analysis lies in their ability to provide a range of values within which the true population parameter is likely to lie, along with a specified level of confidence (such as 95%). Unlike a single point estimate, a confidence interval shows the precision and reliability of an estimate by indicating the possible margin of error. Confidence intervals help in better decision-making, comparison of results, and understanding the uncertainty inherent in sample-based estimates, making them an essential tool in statistical inference.

Ques 8 :-What is the concept of expected value in a probability distribution?

Answer:-The expected value in a probability distribution is the long-run average value of a random variable over many repeated trials. It represents the weighted mean of all possible values of the random variable, where each value is multiplied by its probability and then summed. The expected value indicates the central tendency of a distribution and helps in predicting outcomes, making it a key concept in probability and decision-making under uncertainty.

```python
#Ques 9 :-  Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution.

import numpy as np
import matplotlib.pyplot as plt


data = np.random.normal(loc=50, scale=5, size=1000)


mean_value = np.mean(data)
std_deviation = np.std(data)

print("Mean of the data:", mean_value)
print("Standard Deviation of the data:", std_deviation)


plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Data")
plt.show()
```

```python
#Ques 10 :-  You are working as a data analyst for a retail company. The company has
#collected daily sales data for 2 years and wants you to identify the overall sales trend.
#daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
#235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
#● Explain how you would apply the Central Limit Theorem to estimate the average sales
#with a 95% confidence interval.
#● Write the Python code to compute the mean sales and its confidence interval.

# Import NumPy for numerical calculations
import numpy as np

# Import scipy.stats to get the Z-score for confidence interval
import scipy.stats as stats

# Given daily sales data (sample data)
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert the list into a NumPy array for easier calculations
data = np.array(daily_sales)

# Calculate the sample mean (average daily sales)
mean_sales = np.mean(data)

# Calculate the sample standard deviation
# ddof=1 is used to get the unbiased sample standard deviation
std_dev = np.std(data, ddof=1)

# Calculate the number of observations (sample size)
n = len(data)

# Define the confidence level (95%)
confidence_level = 0.95

# Find the Z-score corresponding to 95% confidence level
# For 95% confidence, Z-score is approximately 1.96
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error using CLT formula
# Margin of Error = Z * (Standard Deviation / √n)
margin_error = z_score * (std_dev / np.sqrt(n))

# Calculate the lower and upper bounds of the confidence interval
lower_bound = mean_sales - margin_error
upper_bound = mean_sales + margin_error

# Print the mean sales value
print("Mean Daily Sales:", mean_sales)

# Print the 95% confidence interval
print("95% Confidence Interval:", (lower_bound, upper_bound))


#Using the Central Limit Theorem, the sample mean is assumed to be normally distributed, which allows us to calculate a 95% confidence interval for the true average daily sales.
```
