Ques 1 :- What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

Answer:- Deep learning is a specialized branch of machine learning that focuses on using artificial neural networks with multiple layers to automatically learn and represent complex patterns in data. Unlike traditional machine learning, which often relies on manual feature engineering where humans must decide which variables or attributes are important, deep learning models can process raw data directly and discover relevant features on their own. The evolution of deep learning can be traced back to the 1950s with the introduction of the perceptron, followed by the development of backpropagation in the 1980s, which allowed multi-layer networks to be trained more effectively. However, progress was limited due to insufficient computational power and small datasets. With the advent of big data and powerful GPUs in the 2000s, researchers were able to train deeper and more complex networks, leading to breakthroughs in areas such as computer vision, speech recognition, and natural language processing. The landmark success of AlexNet in 2012 demonstrated the superiority of deep learning in image classification tasks, sparking widespread adoption across industries. In contrast to traditional machine learning, deep learning requires large amounts of data and significant computational resources but excels at handling unstructured data like images, audio, and text. While traditional models are often more interpretable, deep learning models achieve higher accuracy and performance in complex tasks, though they are frequently criticized for being “black boxes.” Overall, deep learning represents a major leap in artificial intelligence, enabling machines to achieve human-like performance in tasks that were once considered extremely challenging.

Ques 2 :-  Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

Answer:-A Perceptron is the most basic unit of an artificial neural network and was one of the earliest models proposed for machine learning. Its architecture consists of an input layer, a single processing unit (neuron), and an output. Each input feature is multiplied by a corresponding weight, which signifies its importance, and then all weighted inputs are summed together along with a bias term. This total is passed through an activation function — originally a step function — which produces the final output, typically a binary value such as 0 or 1. Functionally, the perceptron acts as a linear classifier, drawing a straight-line decision boundary to separate data into two categories. While this design was groundbreaking in the 1950s, it has notable limitations. The perceptron can only solve problems that are linearly separable, meaning it fails on tasks like the XOR problem where no straight line can divide the classes. Furthermore, the use of a step activation function restricts its ability to learn complex or non-linear relationships, and the model lacks flexibility for handling real-world data. These shortcomings led to the development of multi-layer perceptrons (MLPs) and more advanced neural architectures that incorporate non-linear activation functions, enabling modern deep learning systems to overcome the perceptron’s constraints and tackle far more sophisticated tasks.

Ques 3:- Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

Answer:- In neural networks, the activation function plays a crucial role by introducing non-linearity into the model, allowing it to learn and represent complex patterns beyond simple linear relationships. Without activation functions, a neural network would essentially behave like a linear regression model, regardless of the number of layers, and would fail to capture intricate dependencies in data. Activation functions determine whether a neuron should be “activated” based on the weighted sum of inputs and bias, thereby influencing the flow of information through the network and enabling the network to approximate highly complex functions.

Among the commonly used activation functions, Sigmoid, Tanh, and ReLU each have distinct characteristics. The Sigmoid function maps input values into a range between 0 and 1, making it useful for probability-based outputs, but it suffers from the “vanishing gradient” problem, where gradients become very small for extreme input values, slowing down learning. The Tanh function is similar in shape but maps inputs between -1 and 1, centering the data and often performing better than sigmoid in practice; however, it too faces vanishing gradient issues. The ReLU (Rectified Linear Unit) function, defined as outputting zero for negative inputs and the input itself for positive values, has become the most widely used activation function in deep learning because it is computationally efficient and helps mitigate vanishing gradients. Still, ReLU can suffer from the “dying ReLU” problem, where neurons get stuck outputting zero and stop learning. Overall, activation functions are essential for enabling neural networks to capture non-linear relationships, and the choice among sigmoid, tanh, and ReLU depends on the specific task and architecture, with ReLU generally preferred in modern deep learning due to its simplicity and effectiveness.

Ques 4:- What is the difference between Loss function and Cost function in neural
networks? Provide examples.

Answer:- In neural networks, the loss function and the cost function are closely related but serve slightly different purposes. The loss function measures the error for a single training example — it quantifies how far the predicted output is from the actual target value. For instance, in regression tasks, the Mean Squared Error (MSE) can be used as a loss function, while in classification tasks, Binary Cross-Entropy or Categorical Cross-Entropy are common choices. On the other hand, the cost function represents the overall error across the entire training dataset. It is essentially the average (or sometimes sum) of all individual losses computed by the loss function for each training example. Thus, while the loss function operates at the level of one data point, the cost function aggregates these losses to guide the optimization process during training.

Ques 5:-  What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.

Answer:- In neural networks, optimizers play a vital role in training by adjusting the model’s parameters (weights and biases) to minimize the cost function. They determine how the network learns from data by updating parameters based on the gradients calculated during backpropagation. Without optimizers, the network would not be able to converge toward an accurate solution. Optimizers essentially control the speed and efficiency of learning, balancing between convergence and avoiding issues like overshooting or getting stuck in local minima.

The most common optimizers include Gradient Descent, Adam, and RMSprop, each with distinct characteristics. Gradient Descent is the simplest approach: it updates parameters by moving them in the opposite direction of the gradient of the cost function. While effective, it can be slow and sensitive to the choice of learning rate. RMSprop (Root Mean Square Propagation) improves upon this by adapting the learning rate for each parameter individually, using a moving average of squared gradients to normalize updates. This makes it particularly effective for non-stationary problems and recurrent neural networks. Adam (Adaptive Moment Estimation) combines the benefits of RMSprop and momentum by maintaining both the average of past gradients and the average of squared gradients. This dual adaptation makes Adam highly efficient, robust, and widely used in modern deep learning tasks, as it often converges faster and requires less fine-tuning of hyperparameters. In summary, while gradient descent provides the foundation, RMSprop and Adam offer more sophisticated mechanisms that improve training stability and speed, with Adam being the most popular choice in practice.

```python
# Ques 6 :- Write a Python program to implement a single-layer perceptron from
#scratch using NumPy to solve the logical AND gate.

import numpy as np

# Step activation function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Perceptron class
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        self.weights = np.zeros(input_size + 1)  # +1 for bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, x):
        x_with_bias = np.insert(x, 0, 1)  # add bias input = 1
        weighted_sum = np.dot(self.weights, x_with_bias)
        return step_function(weighted_sum)

    def train(self, X, y):
        for _ in range(self.epochs):
            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                error = target - prediction
                x_with_bias = np.insert(xi, 0, 1)
                self.weights += self.learning_rate * error * x_with_bias

# Training data for AND gate
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 0, 0, 1])  # AND truth table

# Train perceptron
p = Perceptron(input_size=2, learning_rate=0.1, epochs=10)
p.train(X, y)

# Test perceptron
print("Testing AND gate with Perceptron:")
for xi in X:
    print(f"Input: {xi}, Output: {p.predict(xi)}")
```

```python
#Ques 7:- Implement and visualize Sigmoid, ReLU, and Tanh activation functions
#using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-10, 10, 100)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot activation functions
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(3, 1, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Activation Function")
plt.grid(True)

# ReLU
plt.subplot(3, 1, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Activation Function")
plt.grid(True)

# Tanh
plt.subplot(3, 1, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Activation Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Ques 8:-  Use Keras to build and train a simple multilayer neural network on the
#MNIST digits dataset. Print the training accuracy.

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize input data
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Flatten 28x28 images into 784 vector
    Dense(128, activation='relu'),       # Hidden layer with ReLU
    Dense(64, activation='relu'),        # Another hidden layer
    Dense(10, activation='softmax')      # Output layer for 10 classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)

# Print training accuracy
train_acc = history.history['accuracy'][-1]
print(f"Final Training Accuracy: {train_acc:.4f}")
```

```python
# Ques 9:-  Visualize the loss and accuracy curves for a neural network model trained
#on the Fashion MNIST dataset. Interpret the training behavior.

import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalize input data
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build a simple neural network
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32,
                    validation_data=(X_test, y_test), verbose=1)

# Plot loss and accuracy curves
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Ques 10 :-  You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:

● How would you design a deep learning model (perceptron or multilayer NN)?

● Which activation function and loss function would you use, and why?

● How would you train and evaluate the model, considering class imbalance?

● Which optimizer would be suitable, and how would you prevent overfitting?

Answer:- For a real-time fraud detection project in banking, I would design a multilayer neural network (MLP) rather than a single perceptron, since fraud detection involves complex non-linear relationships among structured features like transaction amount, merchant ID, and customer location. The model would include an input layer for normalized features, two to three hidden layers with ReLU activation to capture non-linear dependencies, and a final output layer with a sigmoid activation to produce probabilities for binary classification (fraudulent vs legitimate). For the loss function, binary cross-entropy is most appropriate because it measures the difference between predicted probabilities and actual labels in binary tasks. Training such a model requires careful handling of class imbalance, as fraudulent transactions are rare compared to legitimate ones. To address this, I would use techniques such as oversampling minority classes, undersampling majority classes, or applying class weights in the loss function, while evaluating performance with precision, recall, F1-score, and AUC-ROC rather than accuracy alone. For optimization, Adam is a strong choice because it adapts learning rates dynamically and converges quickly on large datasets. To prevent overfitting, I would incorporate dropout layers, L2 regularization, and early stopping based on validation loss, along with cross-validation to ensure generalization. This workflow ensures the model is efficient enough for real-time inference, robust against imbalance, and capable of minimizing both missed frauds and false alarms.
