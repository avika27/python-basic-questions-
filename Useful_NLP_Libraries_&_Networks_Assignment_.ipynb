Ques 1 :- Compare and contrast NLTK and spaCy in terms of features, ease of use,
and performance.

Answer:- NLTK and spaCy are two popular Python libraries used in Natural Language Processing, but they differ in features, ease of use, and performance. NLTK is mainly designed for education and research purposes and provides a wide range of tools for tasks such as tokenization, stemming, parsing, tagging, and text classification. It includes many linguistic resources and is highly suitable for beginners who want to learn NLP concepts in detail. However, NLTK often requires more code and manual processing steps.

On the other hand, spaCy is designed for industrial and production-level NLP applications. It provides fast and efficient processing with pre-trained models for tasks like named entity recognition, part-of-speech tagging, dependency parsing, and word vectors. spaCy is easier to use for real-world projects because many NLP tasks can be performed with fewer lines of code. In terms of performance, spaCy is significantly faster than NLTK because it is optimized in Cython and built for large-scale text processing, whereas NLTK is comparatively slower and more research-oriented.

In summary, NLTK is best for learning, experimentation, and academic purposes, while spaCy is better suited for high-performance, real-time, and industry-level NLP applications.

Ques 4 :- Describe the architecture and functioning of a Recurrent Natural Network
(RNN).

Answer:- A Recurrent Neural Network (RNN) is a type of artificial neural network specially designed for processing sequential data such as text, speech, and time-series information in Natural Language Processing. Unlike traditional neural networks, an RNN has a memory mechanism that allows information from previous steps to influence the current output. This makes RNNs highly useful for tasks where context and order of data are important, such as language modeling, machine translation, speech recognition, and text generation.

The architecture of an RNN consists of an input layer, hidden layer, and output layer. The key feature is the recurrent connection in the hidden layer, where the output from the previous time step is fed back into the network along with the current input. This creates a loop structure that helps the network remember past information. At each time step, the RNN takes the current input and the previous hidden state to compute a new hidden state and output. Mathematically, the hidden state is updated repeatedly as the sequence progresses, enabling the network to capture temporal dependencies in data.

The functioning of an RNN can be understood using a sentence example. Suppose the sentence is “The cat is sleeping.” The RNN processes one word at a time. While processing the word “sleeping,” the network still remembers previous words like “The,” “cat,” and “is,” helping it understand the context of the sentence. This sequential memory capability is what differentiates RNNs from feedforward neural networks.

However, standard RNNs suffer from problems such as vanishing and exploding gradients when dealing with long sequences, making it difficult to retain long-term dependencies. To solve this issue, advanced variants such as Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) networks were developed. These models improve memory retention and are widely used in modern NLP and deep learning applications.

Ques 5 :- What is the key difference between LSTM and GRU networks in NLP
applications?

Answer:- Long Short-Term Memory (LSTM) and Gated Recurrent Unit (GRU) are advanced types of Recurrent Neural Networks used in Natural Language Processing applications to handle long-term dependencies in sequential data. The key difference between them lies in their architecture and complexity.

LSTM networks contain three gates — input gate, forget gate, and output gate — along with a separate memory cell that stores long-term information. This complex structure allows LSTMs to capture long-range dependencies more effectively, making them highly suitable for tasks such as machine translation, speech recognition, and text generation. However, because of their complex architecture, LSTMs require more computational power and training time.

GRU networks, on the other hand, are a simplified version of LSTM. They use only two gates — update gate and reset gate — and combine the hidden state and memory cell into a single structure. Due to this simpler design, GRUs are faster to train, require fewer parameters, and consume less memory while still achieving performance similar to LSTMs in many NLP tasks.

In summary, LSTMs are more powerful for capturing very long-term dependencies but are computationally heavier, whereas GRUs are simpler, faster, and more efficient for many practical NLP applications.

```python
# Ques 6 :-  Write a Python program using TextBlob to perform sentiment analysis on
#the following paragraph of text:
#“I had a great experience using the new mobile banking app. The interface is intuitive,
#and customer support was quick to resolve my issue. However, the app did crash once
#during a transaction, which was frustrating"
#Your program should print out the polarity and subjectivity scores.

# Install TextBlob (run once if not installed)
!pip install textblob

# Import TextBlob
from textblob import TextBlob

# Sample paragraph
text = """
I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating.
"""

# Create TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

# Print results
print("Sentiment Analysis Results:\n")

print("Polarity Score:", sentiment.polarity)
print("Subjectivity Score:", sentiment.subjectivity)

# Interpretation
if sentiment.polarity > 0:
    print("\nOverall Sentiment: Positive")
elif sentiment.polarity < 0:
    print("\nOverall Sentiment: Negative")
else:
    print("\nOverall Sentiment: Neutral")
```

```python
# Ques 7 :- Given the sample paragraph below, perform string tokenization and
#frequency distribution using Python and NLTK:
#“Natural Language Processing (NLP) is a fascinating field that combines linguistics,
#computer science, and artificial intelligence. It enables machines to understand,
#interpret, and generate human language. Applications of NLP include chatbots,
#sentiment analysis, and machine translation. As technology advances, the role of NLP
#in modern solutions is becoming increasingly critical.”

# Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required tokenizer resources
nltk.download('punkt')
nltk.download('punkt_tab')

# Sample paragraph
text = """
Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.
"""

# Tokenization
tokens = word_tokenize(text)

# Convert tokens to lowercase and remove punctuation
words = [word.lower() for word in tokens if word.isalpha()]

# Frequency Distribution
freq_dist = FreqDist(words)

# Print Tokens
print("Tokens:\n")
print(words)

# Print Frequency Distribution
print("\nFrequency Distribution:\n")
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
# Ques 8 :- Implement a basic LSTM model in Keras for a text classification task using
#the following dummy dataset. Your model should classify sentences as either positive
#(1) or negative (0).
# Dataset
#texts = [
#“I love this project”, #Positive
#“This is an amazing experience”, #Positive
#“I hate waiting in line”, #Negative
#“This is the worst service”, #Negative
#“Absolutely fantastic!” #Positive
#]
#labels = [1, 1, 0, 0, 1]
#Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
#this data. You may use Keras with TensorFlow backend.

# Import required libraries
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
    "I love this project",                 # Positive
    "This is an amazing experience",      # Positive
    "I hate waiting in line",             # Negative
    "This is the worst service",          # Negative
    "Absolutely fantastic!"               # Positive
]

labels = [1, 1, 0, 0, 1]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(texts)

# Vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Pad sequences
max_length = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(
    sequences,
    maxlen=max_length,
    padding='post'
)

# Convert labels to numpy array
labels = np.array(labels)

# Print processed data
print("Word Index:\n")
print(tokenizer.word_index)

print("\nPadded Sequences:\n")
print(padded_sequences)

# Build LSTM model
model = Sequential()

# Embedding Layer
model.add(Embedding(
    input_dim=vocab_size,
    output_dim=16,
    input_length=max_length
))

# LSTM Layer
model.add(LSTM(32))

# Output Layer
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

# Model Summary
print("\nModel Summary:\n")
model.summary()

# Train model
model.fit(
    padded_sequences,
    labels,
    epochs=20,
    verbose=1
)

# Test Prediction
test_sentence = ["I really love this service"]

# Preprocess test sentence
test_seq = tokenizer.texts_to_sequences(test_sentence)
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')

# Predict sentiment
prediction = model.predict(test_pad)

print("\nPrediction Score:", prediction[0][0])

if prediction[0][0] > 0.5:
    print("Predicted Sentiment: Positive")
else:
    print("Predicted Sentiment: Negative")
```

```python
# Ques 10 :- You are working on a chatbot for a mental health platform. Explain how
#you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
#NLP to understand and respond to user input effectively. Detail your architecture, data
#preprocessing pipeline, and any ethical considerations.

# Mental Health Chatbot using spaCy + GRU/LSTM
# ------------------------------------------------

# Install required libraries (Run once)
!pip install spacy tensorflow nltk

# Download spaCy English model
!python -m spacy download en_core_web_sm

# Import libraries
import numpy as np
import spacy
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# ------------------------------------------------
# Sample Mental Health Dataset
# ------------------------------------------------

texts = [
    "I feel very happy today",
    "I am stressed about my exams",
    "I feel lonely and depressed",
    "Life is beautiful and amazing",
    "I am anxious and worried",
    "I feel calm and relaxed"
]

# Labels
# 1 = Positive Mental State
# 0 = Negative Mental State

labels = [1, 0, 0, 1, 0, 1]

# ------------------------------------------------
# Text Preprocessing using spaCy
# ------------------------------------------------

processed_texts = []

for text in texts:

    # Process text with spaCy
    doc = nlp(text.lower())

    # Tokenization + Lemmatization
    tokens = [
        token.lemma_
        for token in doc
        if not token.is_stop and token.is_alpha
    ]

    processed_texts.append(" ".join(tokens))

print("Processed Text:\n")
print(processed_texts)

# ------------------------------------------------
# Tokenization using Keras
# ------------------------------------------------

tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(processed_texts)

# Vocabulary size
vocab_size = len(tokenizer.word_index) + 1

# Pad sequences
max_length = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(
    sequences,
    maxlen=max_length,
    padding='post'
)

# Convert labels
labels = np.array(labels)

# ------------------------------------------------
# Build LSTM Model
# ------------------------------------------------

model = Sequential()

# Embedding Layer
model.add(Embedding(
    input_dim=vocab_size,
    output_dim=16,
    input_length=max_length
))

# LSTM Layer
model.add(LSTM(32))

# Output Layer
model.add(Dense(1, activation='sigmoid'))

# Compile Model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model Summary
print("\nMODEL SUMMARY:\n")
model.summary()

# ------------------------------------------------
# Train Model
# ------------------------------------------------

model.fit(
    padded_sequences,
    labels,
    epochs=20,
    verbose=1
)

# ------------------------------------------------
# Test Chatbot Sentiment Prediction
# ------------------------------------------------

test_text = ["I feel stressed and lonely"]

# spaCy preprocessing
doc = nlp(test_text[0].lower())

tokens = [
    token.lemma_
    for token in doc
    if not token.is_stop and token.is_alpha
]

clean_text = " ".join(tokens)

# Convert to sequence
test_seq = tokenizer.texts_to_sequences([clean_text])

# Pad sequence
test_pad = pad_sequences(
    test_seq,
    maxlen=max_length,
    padding='post'
)

# Predict
prediction = model.predict(test_pad)

print("\nPrediction Score:", prediction[0][0])

# Generate Response
if prediction[0][0] > 0.5:
    print("Chatbot Response: I'm glad you're feeling positive today.")
else:
    print("Chatbot Response: I'm sorry you're feeling this way. Please remember that support is available.")
```
