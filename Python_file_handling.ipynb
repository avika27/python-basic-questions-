Ques 1:- What is the difference between interpreted and compiled languages?

Answer:- Compiled languages translate the entire source code into machine code before execution, which makes them faster and more efficient because the computer runs the pre-compiled executable directly. Errors are detected at compile time, and the program must be recompiled after every change. Examples include C, C++ and Go. In contrast, interpreted languages are executed line-by-line by an interpreter at runtime, which makes them slower but easier to test and debug since no separate compilation step is required. Errors appear during execution, and the same code can run on any system that has the required interpreter. Python, JavaScript and Ruby are common examples of interpreted languages.

Ques 2:- What is exception handling in Python?

Answer:-
Exception handling in Python is a mechanism used to manage and handle errors that occur while a program is running, so the program does not stop unexpectedly. Python uses the try–except block to catch these runtime errors (called exceptions) and handle them safely. The code that might cause an error is written inside the try block, and if an exception occurs, the except block runs instead of the program crashing. This allows the program to continue running, show meaningful error messages, or take alternative actions. In simple terms, exception handling helps make Python programs more stable, safe, and error-tolerant.

Ques 9:-What is the difference between multithreading and multiprocessing?

Answer:-Multithreading uses multiple threads within the same process. All threads share the same memory space, which makes communication easier but also creates limitations like the Global Interpreter Lock (GIL) in CPython, which prevents true parallel CPU execution. Multithreading is best for tasks that involve waiting—such as file I/O, network requests, or user input—because threads can run concurrently without heavy CPU use.

Multiprocessing, on the other hand, uses multiple separate processes, each with its own memory space and its own Python interpreter. This bypasses the GIL and allows true parallel execution on multiple CPU cores. It is ideal for CPU-intensive tasks like computations, data processing, and mathematical operations. However, communication between processes is slower and more complex because memory is not shared.

Ques 11:- What is memory management in Python?

Python automatically handles memory through a private heap, where all objects and data structures are stored. Ques 15:- How does Python's garbage collection system work ?

Answer:-Python’s garbage collection system works by automatically detecting and removing objects that are no longer needed, so memory can be reused. It mainly uses two mechanisms:

Reference Counting:
Every object in Python keeps a count of how many references are pointing to it.

When a new reference is created, the count increases.

When a reference is removed, the count decreases.

When the count becomes zero, the object is immediately deleted, and its memory is freed.

Garbage Collector for Circular References:
Sometimes objects reference each other in a loop (circular reference), making reference counting alone insufficient.
Python’s built-in garbage collector (in the gc module) periodically scans memory to find such unused circular references and frees them safely.

Together, these features help Python manage memory efficiently without requiring the programmer to manually allocate or deallocate memory.

Ques 18:-What is the difference between os.fork() and multiprocessing in Python ?

Answer:-The difference between os.fork() and multiprocessing in Python lies in how they create and manage new processes:

os.fork() is a low-level system call available only on Unix/Linux. It creates a child process by duplicating the current process. Both parent and child continue running the same code from the point of the fork. It gives you direct control over process creation, but it is platform-dependent and harder to manage. You must handle communication, synchronization, and resource sharing manually.

Multiprocessing, on the other hand, is a high-level Python module that works on all major platforms, including Windows. It provides an easy and safer way to create and manage multiple processes. The module handles process creation, inter-process communication, shared memory, queues, locks, and pools internally. It avoids issues like the Global Interpreter Lock (GIL) and allows true parallel execution.

Ques 23:- What are the challenges associated with memory management in Python?

Answer:-Memory management in Python faces several challenges because of its automatic and dynamic nature:

High memory usage:
Python objects and data structures often consume more memory compared to lower-level languages like C, especially due to dynamic typing and object overhead.

Garbage collection overhead:
The garbage collector may pause the program temporarily to clean up unused objects, which can affect performance in memory-intensive applications.

Circular references:
Objects referencing each other can create cycles that reference counting alone cannot handle. Although the garbage collector manages these, it adds complexity and extra work.

Fragmentation:
Frequent allocation and deallocation of memory can lead to fragmentation, where free memory becomes scattered, reducing efficiency.

Unintentional memory leaks:
Poor programming practices—such as keeping unnecessary references—can prevent objects from being collected, causing memory usage to grow over time.

Difficulty in manual control:
Python abstracts memory handling, which is convenient, but offers less manual control for developers who need fine-tuned memory optimization.

```python
# Ques 14:-Write a program that uses the logging module to log both informational and error messages

import logging


logging.basicConfig(
    filename="app.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def divide(a, b):
    try:
        result = a / b
        logging.info(f"Division successful: {a} / {b} = {result}")
        return result

    except ZeroDivisionError:
        logging.error("Error: Attempted division by zero!")
        return None

    except Exception as e:
        logging.error(f"Unexpected error: {e}")
        return None

divide(10, 2)
divide(5, 0)
divide("x", 3)
```

```python
# Ques 18:- How would you implement a basic logging setup that logs to a file with rotation after 1MB?

import logging
from logging.handlers import RotatingFileHandler


logger = logging.getLogger("my_app")
logger.setLevel(logging.INFO)
log_file = "app.log"
handler = RotatingFileHandler(filename=log_file,
                              mode="a",
                              maxBytes=1 * 1024 * 1024,
                              backupCount=5,
                              encoding="utf-8",
                              delay=False)


formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(name)s - %(message)s")
handler.setFormatter(formatter)


logger.addHandler(handler)


def divide(a, b):
    try:
        result = a / b
        logger.info("Division: %s / %s = %s", a, b, result)
        return result
    except ZeroDivisionError:
        logger.error("Attempted division by zero: %s / %s", a, b, exc_info=True)
        return None
    except Exception:
        logger.exception("Unexpected error in divide()")
        return None

if __name__ == "__main__":

    for i in range(100000):
        logger.info("Log line %d: some repetitive info to grow the log", i)

    divide(10, 2)
    divide(5, 0)

```

```python
# Ques 21:-Write a Python program that reads a file and prints the number of occurrences of a specific word.

def count_word(filename, word):
    try:
        with open(filename, "r") as file:
            content = file.read().lower()
            word = word.lower()

            count = content.split().count(word)
            print(f"The word '{word}' occurs {count} times in the file.")

    except FileNotFoundError:
        print("Error: The file does not exist!")

    except Exception as e:
        print("An unexpected error occurred:", e)



count_word("sample.txt", "python")
```

```python
# Ques 23:- Write a Python program that writes to a log file when an error occurs during file handling.

import logging


logging.basicConfig(
    filename="file_errors.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def read_file(filename):
    try:
        with open(filename, "r") as file:
            content = file.read()
            print("File content:\n", content)

    except FileNotFoundError:
        logging.error(f"File not found: {filename}")
        print("Error: The file does not exist!")

    except PermissionError:
        logging.error(f"Permission denied while accessing: {filename}")
        print("Error: You don't have permission to read this file!")

    except Exception as e:
        logging.error(f"Unexpected error while reading {filename}: {e}")
        print("An unexpected error occurred!")

read_file("sample.txt")
```
